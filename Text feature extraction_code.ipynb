# 2. Preprocessing the text by removing stop words, converting all text to lowercase, and removing special characters.

```python
import string
```

```python
def rem_punct(mystr):
    return mystr.translate(str.maketrans('','',string.punctuation)) # removed special characters by using rem_punct function
```

```python
# creating a clean_text column to show the difference
```

```python
df['clean_text'] = df['headline_text'].apply(rem_punct)   
```

```python
df
```

```python
df['clean_text'] = df['clean_text'].apply(str.lower) # converted all text to lowercase
```

```python
df
```

```python
# removing stop words
# created a clean_text_sw column to differentiate from the previous clean_text
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
def remove_stopwords(text):
    doc = nlp(text)
    tokens = [token.text for token in doc if not token.is_stop]
    return ' '.join(tokens)
df['clean_text_sw'] = df['clean_text'].apply(remove_stopwords)
```

```python
df
```

# 3. using Scikkit-Learn's CountVectorizer to convert the preprocessed text into a matrix of token counts.

```python
#Scikkit-Learn's CountVectorizer

from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vect = CountVectorizer()
```

```python
X_train_counts = count_vect.fit_transform(df.clean_text_sw)
```

```python
X_train_counts.T.todense()
```

# 4. Calculate the TF-IDF (Term Frequency-Inverse Document Frequency) score for each word in the dataset using Scikit-learn

```python
#Transform Counts to Frequencies with Tf-idf
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(df.clean_text_sw)
```

```python
pd.DataFrame(tfidf_matrix.T.todense())
```

# 5. Converting the TF-IDF scores into a Pandas dataframe and display the top 10 words with the highest scores.

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
tfidf_df
```

```python
tfidf_df_mean = tfidf_df.mean()
```

```python
tfidf_df_mean
```

```python
top_words = tfidf_df_mean.sort_values(ascending=False).head(10)
```

```python
print(top_words)
```
